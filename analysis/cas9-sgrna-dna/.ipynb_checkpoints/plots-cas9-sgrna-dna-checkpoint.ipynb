# RMSD

```python
RMSD_pro = np.loadtxt('data/rmsd_cas9-sgrna-dna.proa.txt')
RMSD_dna = np.loadtxt('data/rmsd_cas9-sgrna-dna.dnaa-b.txt')
RMSD_rna = np.loadtxt('data/rmsd_cas9-sgrna-dna.rnaa.txt')
```

```python
plt.figure(figsize=(4.25,3.25))
plt.plot(RMSD_pro[:,0],RMSD_pro[:,1], 'b', label='Cas9')
plt.plot(RMSD_pro[:,0],RMSD_dna[:,1], 'g', label='DNA')
plt.plot(RMSD_pro[:,0],RMSD_rna[:,1], 'r', label='sgRNA')
#--------------------------------------------------------------------------#
plt.xlabel('Simulation Time(ns)', family="Arial", size='10', weight='bold')
plt.ylabel('RMSD ($\AA$)', family="Arial", size='10', weight='bold')
#--------------------------------------------------------------------------#
plt.xticks(family="Arial", size='10', weight='bold')
plt.yticks(family="Arial", size='10', weight='bold')
plt.tick_params(axis='y', which='both', left='on', right='on')
# plt.margins(0.025)
# plt.grid(axis='y', linestyle='dotted')
font = font_manager.FontProperties(family="Arial", size='9', weight='bold')
plt.legend(loc='lower right', ncol=3, prop=font)#bbox_to_anchor=(0.1, 1),, fontname="Arial")
plt.tight_layout()#rect=(-0.025,0,1.025,1.025))
#--------------------------------------------------------------------------#
plt.savefig('./figures/RMSD.png', dpi=300)
plt.show()
```

# DIPOLE MOMENT

```python
dp_n = np.loadtxt('data/dipole_nucleobase_cas9-sgrna-dna-rnaa.txt', skiprows=1)

Z = np.zeros((200,96))

for i in range(len(dp_n[:,0])):
    j = int((dp_n[i,0]/100) - 1)
    k = int(dp_n[i,1] - dp_n[0,1])
    Z[j,k] = dp_n[i,6]

R = []
for r in range(len(Z[0,:])):
    avg = np.mean(Z[:,r])
    std = np.std(Z[:,r])
    R.append((r+dp_n[0,1],avg,std))
R = np.array(R)

plt.figure(figsize=(6.5,2.5))
plt.bar(np.arange(len(R[1:-1,0])),R[1:-1,1],yerr=R[1:-1,2])
plt.xticks(np.arange(0,len(R[1:-1,0]),10),np.arange(int(R[1,0]),int(R[-2,0])+1,10))
plt.ylabel('Dipole: nucleobase (Debyes)', fontweight='bold',fontname="Arial")
plt.xlabel('Residue Number', fontweight='bold',fontname="Arial")
plt.xticks(fontname="Arial")
plt.yticks(fontname="Arial")
plt.grid(axis='y')
font = font_manager.FontProperties(family='Arial')
# plt.legend(ncol=3, prop=font)#, fontname="Arial")
plt.tight_layout()
plt.savefig('./figures/dipole_nucleobase_cas9-sgrna-dna-rnaa.png', dpi=300)
plt.show()
```

# DIPOLE MOMENT DISTRIBUTION

```python
# Amino Acids Side Chains
dp = []

resnames = ['ALA', 'ARG', 'ASN', 'ASP', 'CYS', 'GLN', 'GLU', 'HSD', 'ILE', 'LEU', 'LYS', 'MET', 'PHE', 'PRO', 'SER', 'THR', 'TRP', 'TYR', 'VAL']


resnames = ['ALA','ILE', 'LEU', 'PHE', 'PRO', 'VAL']
for res in resnames:
    print(res)
    df = pd.read_csv('data/aasc_dipole_moments/%s.txt' % (res), sep=' ')
    df.dropna(inplace=True)
    df = df[(df != 0).all(1)]
    d = df.to_numpy()
    m = np.mean(d[:,3])
    s = np.std(d[:,3])
    df2 = pd.read_csv('data/aasc_dipole_moments/c36.%s.txt' % (res), sep=' ')
    df2.dropna(inplace=True)
    df2 = df2[(df2 != 0).all(1)]
    d2 = df2.to_numpy()
    m2 = np.mean(d2[:,3])
    s2 = np.std(d2[:,3])
    dp.append((res,m,s))
    plt.figure(figsize=(1.4,2.0))
    plt.title("%s\n$\mu$ = %5.3f\n$\sigma$ = %5.3f" % (res,m,s), family="Arial", size='8', weight='bold')
    plt.ylim((0,2))
    plt.hist(d[:,3], bins=50, range=(max(0,m-5*s),m+5*s), density=True, stacked=True, orientation='horizontal', color='xkcd:green')
    plt.hist(d2[:,3], bins=50, range=(max(0,m2-5*s2),m2+5*s2), density=True, stacked=True, orientation='horizontal', color='xkcd:black', alpha=0.4)
    #plt.hist(d[:,3], bins=100, range=(0,2), orientation='horizontal', color='xkcd:green')
    #plt.hist(d2[:,3], bins=100, range=(0,2), orientation='horizontal', color='xkcd:black', alpha=0.4)
    #plt.hist(d[:,3], bins=50, range=(0,15), density=True)
    plt.ylabel('Dipole Moment (Debyes)', family="Arial", size='8', weight='bold')
    plt.xticks(family="Arial", size='8', weight='bold')
    plt.yticks([0,1,2], family="Arial", size='8', weight='bold')
    plt.tight_layout(rect=(0,0,1.075,1.075))
    #plt.tight_layout()
    plt.savefig('figures/%s.aasc_dp.png' % (res), dpi=300)
    plt.close()

resnames = ['CYS', 'MET', 'SER', 'THR', 'TRP', 'TYR']
for res in resnames:
    print(res)
    df = pd.read_csv('data/aasc_dipole_moments/%s.txt' % (res), sep=' ')
    df.dropna(inplace=True)
    df = df[(df != 0).all(1)]
    d = df.to_numpy()
    m = np.mean(d[:,3])
    s = np.std(d[:,3])
    df2 = pd.read_csv('data/aasc_dipole_moments/c36.%s.txt' % (res), sep=' ')
    df2.dropna(inplace=True)
    df2 = df2[(df2 != 0).all(1)]
    d2 = df2.to_numpy()
    m2 = np.mean(d2[:,3])
    s2 = np.std(d2[:,3])
    dp.append((res,m,s))
    plt.figure(figsize=(1.4,2.0))
    plt.title("%s\n$\mu$ = %5.3f\n$\sigma$ = %5.3f" % (res,m,s), family="Arial", size='8', weight='bold')
    plt.ylim((0,2))
    plt.hist(d[:,3], bins=50, range=(max(0,m-5*s),m+5*s), density=True, stacked=True, orientation='horizontal', color='xkcd:green')
    plt.hist(d2[:,3], bins=50, range=(max(0,m2-5*s2),m2+5*s2), density=True, stacked=True, orientation='horizontal', color='xkcd:black', alpha=0.4)
    #plt.hist(d[:,3], bins=50, range=(0,15), density=True)
    plt.ylabel('Dipole Moment (Debyes)', family="Arial", size='8', weight='bold')
    plt.xticks(family="Arial", size='8', weight='bold')
    plt.yticks([0,1,2,3,4,5], family="Arial", size='8', weight='bold')
    plt.tight_layout(rect=(0,0,1.075,1.075))
    #plt.tight_layout()
    plt.savefig('figures/%s.aasc_dp.png' % (res), dpi=300)
    plt.close()

resnames = ['ASN', 'ASP', 'GLN', 'HSD']
for res in resnames:
    print(res)
    df = pd.read_csv('data/aasc_dipole_moments/%s.txt' % (res), sep=' ')
    df.dropna(inplace=True)
    df = df[(df != 0).all(1)]
    d = df.to_numpy()
    m = np.mean(d[:,3])
    s = np.std(d[:,3])
    df2 = pd.read_csv('data/aasc_dipole_moments/c36.%s.txt' % (res), sep=' ')
    df2.dropna(inplace=True)
    df2 = df2[(df2 != 0).all(1)]
    d2 = df2.to_numpy()
    m2 = np.mean(d2[:,3])
    s2 = np.std(d2[:,3])
    dp.append((res,m,s))
    plt.figure(figsize=(1.2,2.0))
    plt.title("%s\n$\mu$ = %5.3f\n$\sigma$ = %5.3f" % (res,m,s), family="Arial", size='8', weight='bold')
    plt.ylim((3,8))
    plt.hist(d[:,3], bins=50, range=(max(0,m-5*s),m+5*s), density=True, stacked=True, orientation='horizontal', color='xkcd:green')
    plt.hist(d2[:,3], bins=50, range=(max(0,m2-5*s2),m2+5*s2), density=True, stacked=True, orientation='horizontal', color='xkcd:black', alpha=0.4)
    #plt.hist(d[:,3], bins=50, range=(0,15), density=True)
    plt.ylabel('Dipole Moment (Debyes)', family="Arial", size='8', weight='bold')
    plt.xticks(family="Arial", size='8', weight='bold')
    plt.yticks([3,4,5,6,7,8], family="Arial", size='8', weight='bold')
    plt.tight_layout(rect=(0,0,1.075,1.075))
    #plt.tight_layout()
    plt.savefig('figures/%s.aasc_dp.png' % (res), dpi=300)
    plt.close()

resnames = ['ARG', 'GLU', 'LYS']
for res in resnames:
    print(res)
    df = pd.read_csv('data/aasc_dipole_moments/%s.txt' % (res), sep=' ')
    df.dropna(inplace=True)
    df = df[(df != 0).all(1)]
    d = df.to_numpy()
    m = np.mean(d[:,3])
    s = np.std(d[:,3])
    df2 = pd.read_csv('data/aasc_dipole_moments/c36.%s.txt' % (res), sep=' ')
    df2.dropna(inplace=True)
    df2 = df2[(df2 != 0).all(1)]
    d2 = df2.to_numpy()
    m2 = np.mean(d2[:,3])
    s2 = np.std(d2[:,3])
    dp.append((res,m,s))
    plt.figure(figsize=(1.2,2.0))
    plt.title("%s\n$\mu$ = %5.3f\n$\sigma$ = %5.3f" % (res,m,s), family="Arial", size='8', weight='bold')
    plt.ylim((4,14))
    plt.hist(d[:,3], bins=50, range=(max(0,m-5*s),m+5*s), density=True, stacked=True, orientation='horizontal', color='xkcd:green')
    plt.hist(d2[:,3], bins=50, range=(max(0,m2-5*s2),m2+5*s2), density=True, stacked=True, orientation='horizontal', color='xkcd:black', alpha=0.4)
    #plt.hist(d[:,3], bins=50, range=(0,15), density=True)
    plt.ylabel('Dipole Moment (Debyes)', family="Arial", size='8', weight='bold')
    plt.xticks(family="Arial", size='8', weight='bold')
    plt.yticks([4,6,8,10,12,14], family="Arial", size='8', weight='bold')
    plt.tight_layout(rect=(0,0,1.075,1.075))
    #plt.tight_layout()
    plt.savefig('figures/%s.aasc_dp.png' % (res), dpi=300)
    plt.close()

dp = np.array(dp)
np.savetxt('dipole_aasc.txt', dp, fmt="%s")

!ls figures |grep aasc | wc -l
```
