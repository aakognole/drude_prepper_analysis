```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
plt.rc('axes', axisbelow=True)
```

```python
!mkdir -p figures
```

# RMSD

```python
RMSD_proa = np.loadtxt('data/rmsd_backbone_fit_to_equil.xvg', comments=('#','@'))

plt.figure(figsize=(4.5,3.5))
plt.plot(RMSD_proa[:,0],10*RMSD_proa[:,1], 'b', label='Ubiquitin')
#--------------------------------------------------------------------------#
plt.xlabel('Simulation Time(ns)', family="Arial", size='10', weight='bold')
plt.ylabel('RMSD ($\AA$)', family="Arial", size='10', weight='bold')
#--------------------------------------------------------------------------#
plt.xticks(family="Arial", size='10', weight='bold')
plt.yticks(family="Arial", size='10', weight='bold')
plt.tick_params(axis='y', which='both', left='on', right='on')
# plt.margins(0.025)
plt.grid(axis='y', linestyle='dotted')
font = font_manager.FontProperties(family="Arial", size='9', weight='bold')
plt.legend(loc='lower right', ncol=3, prop=font)#bbox_to_anchor=(0.1, 1),, fontname="Arial")
plt.tight_layout()#rect=(-0.025,0,1.025,1.025))
#--------------------------------------------------------------------------#
plt.savefig('./figures/RMSD.png', dpi=300)
plt.show()
```

# RMSF

```python
RMSF_proa = np.loadtxt('data/rmsf_backbone_aver_over_res.xvg', comments=('#','@'))

plt.figure(figsize=(4.5,2))
plt.plot(RMSF_proa[:,0],RMSF_proa[:,1]*10, 'b', label='Ubiquitin')
plt.xlabel('Residue Number', family="Arial", size='10', weight='bold')
plt.ylabel('RMSF ($\AA$)', family="Arial", size='10', weight='bold')
plt.xticks(family="Arial", size='10', weight='bold')
plt.yticks(family="Arial", size='10', weight='bold')
plt.tick_params(axis='y', which='both', left='on', right='on')
# plt.margins(0.025)
plt.grid(axis='y', linestyle='dotted')
font = font_manager.FontProperties(family="Arial", size='9', weight='bold')
plt.legend(loc='upper center', ncol=3, prop=font)#bbox_to_anchor=(0.1, 1),, fontname="Arial")
plt.tight_layout()#rect=(-0.025,0,1.025,1.025))
#--------------------------------------------------------------------------#
plt.savefig('./figures/RMSF_proa.png', dpi=300)
plt.show()
```
