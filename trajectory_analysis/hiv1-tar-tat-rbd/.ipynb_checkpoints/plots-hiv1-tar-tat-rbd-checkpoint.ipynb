```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
plt.rc('axes', axisbelow=True)
import MDAnalysis as mda
```

```python
mkdir -p figures
```

# RMSD

```python
RMSD_proa = np.loadtxt('data/rmsd_hiv1-tar-rbd.proa.txt')
RMSD_rnaa = np.loadtxt('data/rmsd_hiv1-tar-rbd.rnaa.txt')
```

```python
plt.figure(figsize=(4.5,3.5))
plt.plot(RMSD_proa[:,0],RMSD_proa[:,1], 'b', label='TAT RBD')
plt.plot(RMSD_proa[:,0],RMSD_rnaa[:,1], 'r', label='TAR RNA')
#--------------------------------------------------------------------------#
plt.xlabel('Simulation Time(ns)', family="Arial", size='10', weight='bold')
plt.ylabel('RMSD ($\AA$)', family="Arial", size='10', weight='bold')
#--------------------------------------------------------------------------#
plt.xticks(family="Arial", size='10', weight='bold')
plt.yticks(family="Arial", size='10', weight='bold')
plt.tick_params(axis='y', which='both', left='on', right='on')
# plt.margins(0.025)
# plt.grid(axis='y', linestyle='dotted')
font = font_manager.FontProperties(family="Arial", size='9', weight='bold')
plt.legend(loc='lower right', ncol=3, prop=font)#bbox_to_anchor=(0.1, 1),, fontname="Arial")
plt.tight_layout()#rect=(-0.025,0,1.025,1.025))
#--------------------------------------------------------------------------#
plt.savefig('./figures/RMSD.png', dpi=300)
plt.show()
```

# DIPOLE MOMENT

```python
u = mda.Universe('step2_drude.pdb')
rna = u.select_atoms('nucleic')
rn = []
col = []
for res in rna.residues[1:-1]:
    rn.append((res.resid,str(res.resname)[0]))
    if str(res.resname)[0] == 'G':
        col.append('g')
    if str(res.resname)[0] == 'A':
        col.append('xkcd:yellow')
    if str(res.resname)[0] == 'C':
        col.append('r')
    if str(res.resname)[0] == 'U':
        col.append('xkcd:cyan')
rn = np.array(rn)
# col = np.array(col)
print(rn)
print(col)

seq = ' '
for res in rna.residues[1:-1]:
    seq = seq + str(res.resname)[0]
print(seq)
```

```python
dp_n = np.loadtxt('data/dipole_nucleobase_hiv1-tar-rbd-rnaa.txt', skiprows=1)

Z = np.zeros((200,31))

for i in range(len(dp_n[:,0])):
    j = int((dp_n[i,0]/100) - 1)
    k = int(dp_n[i,1] - dp_n[0,1])
    Z[j,k] = dp_n[i,6]

R = []
for r in range(len(Z[0,:])):
    avg = np.mean(Z[:,r])
    std = np.std(Z[:,r])
    R.append((r+dp_n[0,1],avg,std))
R = np.array(R)
# R = np.delete(R,8,0)
print(R)


plt.figure(figsize=(4.5,3.5))
plt.bar(np.arange(len(R[1:-1,0])),R[1:-1,1],yerr=R[1:-1,2],color=['g', 'g', 'r', 'xkcd:yellow', 'g', 'xkcd:yellow', 'xkcd:cyan','g', 'xkcd:cyan', 'g', 'xkcd:yellow', 'g', 'r', 'r', 'xkcd:cyan', 'g', 'g', 'g', 'xkcd:yellow', 'g', 'r', 'xkcd:cyan', 'r', 'xkcd:cyan', 'r', 'xkcd:cyan', 'g', 'r', 'r'])
# plt.bar(rn[:],R[1:-1,1],yerr=R[1:-1,2])
plt.xticks(np.arange(0,len(R[1:-1,0]),4),np.arange(int(R[1,0]),int(R[-2,0])+1,4))
plt.ylim((0,13))
plt.ylabel('Dipole: nucleobase (Debyes)', fontweight='bold',fontname="Arial")
plt.xlabel('Residue Number', fontweight='bold',fontname="Arial")
#--------------------------------------------------------------------------#
# plt.xticks(np.arange(len(rn[:,0])),rn[:,0],family="Arial", size='10', weight='bold')
# plt.xticks(family="Arial", size='10', weight='bold')
plt.yticks(family="Arial", size='10', weight='bold')
plt.tick_params(axis='y', which='both', left='on', right='on')
plt.text(-0.5,12.2,'GGCAGAU  UGAGCCUGGGAGCUCUCUGCC', family='Arial', size='11.6')
# plt.margins(0.025)
plt.grid(axis='y', linestyle='dotted')
font = font_manager.FontProperties(family="Arial", size='9', weight='bold')
# plt.legend(loc='lower right', ncol=3, prop=font)#bbox_to_anchor=(0.1, 1),, fontname="Arial")
plt.tight_layout()#rect=(-0.025,0,1.025,1.025))
#--------------------------------------------------------------------------#
plt.savefig('./figures/dipole_nucleobase_hiv1-tar-rbd-rnaa-color.png', dpi=300)
plt.show()
```
